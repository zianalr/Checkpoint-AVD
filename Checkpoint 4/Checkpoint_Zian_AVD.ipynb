###**Outliers Values**

Pada tahapan ini, akan dilakukan pengecekan terhadap dataset untuk mengetahui apakah ada data yang bernilai ekstrem (Outliers Value)

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

## **Outliers**

Pada tahap ini, akan dilakukan penanganan terhadap data yang bersifat ekstrem (Outliers Value), agar tidak ada kekeliruan data dan memudahkan untuk melanjutkan tahap analisis

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

```python
columns_to_impute = ["duration_ms"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

Variabel yang tersedia lebih banyak berupa kategori seperti genre dan artist, serta variabel numerik seperti popularity dan duration_ms. Oleh karena itu, penggunaan line chart menjadi kurang relevan karena tidak menunjukkan perubahan data dari waktu ke waktu.

Selain itu, apabila menggunakan variabel duration_ms sebagai acuan dalam line chart, jumlah data yang cukup banyak akan membuat titik-titik data menjadi sangat rapat atau berdempetan pada sumbu grafik. Kondisi ini menyebabkan visualisasi menjadi sulit dibaca dan tidak memberikan informasi yang jelas mengenai pola data. Dengan demikian, visualisasi lain seperti histogram, box plot, bar chart, scatter plot, atau heatmap lebih sesuai digunakan karena mampu menampilkan distribusi data, perbandingan antar kategori, serta hubungan antar variabel secara lebih jelas tanpa memerlukan dimensi waktu.

## **Histogram**

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['popularity'], bins=20, kde=True)
plt.title('Distribusi Popularitas Lagu')
plt.xlabel('Popularitas')
plt.ylabel('Frekuensi')
plt.show()
```

Histogram menunjukkan bahwa sebagian besar lagu memiliki tingkat popularitas pada rentang menengah, dengan jumlah lagu yang sangat populer atau sangat rendah relatif lebih sedikit. Distribusi ini mengindikasikan bahwa hanya sebagian kecil lagu yang benar-benar mencapai tingkat popularitas tinggi.

Berdasarkan distribusi tersebut, industri musik dapat lebih fokus pada strategi promosi dan pemasaran untuk meningkatkan peluang lagu mencapai tingkat popularitas yang lebih tinggi. Optimalisasi algoritma rekomendasi, promosi di media sosial, serta kolaborasi dengan artis lain dapat membantu meningkatkan visibilitas lagu sehingga peluang menjadi populer semakin besar.

## **Box Plot**

```python
VARIABEL1 = df['genre'].value_counts().head(5).index
VARIABEL2 = df[df['genre'].isin(VARIABEL1)]

plt.figure(figsize=(12, 7)) # Ukuran figure
sns.boxplot(x='genre', y='popularity', data=VARIABEL2, palette='RdPu', order=VARIABEL1, hue='genre', legend=False)
plt.title('Perbandingan Popularitas Lagu Berdasarkan Genre', fontsize=16) # Judul
plt.xlabel('Genre', fontsize=12) # X Keterangan
plt.ylabel('Popularitas', fontsize=12) # Y Keterangan
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Box plot menunjukkan bahwa setiap genre memiliki sebaran popularitas yang berbeda. Beberapa genre seperti ambient dan alternative memiliki median popularitas yang relatif lebih tinggi dibandingkan genre lainnya, sementara genre seperti acoustic dan afrobeat memiliki variasi yang lebih besar dalam tingkat popularitas.

Berdasarkan perbedaan sebaran tersebut, pelaku industri musik dapat melakukan analisis lebih lanjut terhadap karakteristik audiens setiap genre. Genre dengan popularitas yang lebih stabil dapat dijadikan fokus dalam strategi promosi, sementara genre dengan variasi tinggi dapat dieksplorasi melalui inovasi musik atau kolaborasi untuk meningkatkan konsistensi popularitas lagu.

## **ScatterPlot**

```python
plt.figure(figsize=(10, 6)) # Ukuran
plt.scatter(df['duration_ms'], df['popularity'], alpha=0.7, color='pink') # Color (Warna)
plt.title('Hubungan Durasi Lagu dan Popularitas Berdasarkan Genre', fontsize=16) # Judul
plt.xlabel('Durasi(detik)', fontsize=12) # X Keterangan
plt.ylabel('Popularitas', fontsize=12) # Y Keterangan
plt.grid(True)
plt.tight_layout()
plt.show()
```

Scatter plot menunjukkan bahwa tidak terdapat hubungan yang jelas antara durasi lagu dan tingkat popularitas. Lagu dengan durasi pendek maupun panjang dapat memiliki tingkat popularitas yang beragam.

Berdasarkan pola tersebut, durasi lagu tidak perlu menjadi faktor utama dalam menentukan strategi produksi musik. Produsen dan artis dapat lebih berfokus pada kualitas komposisi, lirik, serta strategi distribusi dan promosi untuk meningkatkan daya tarik lagu bagi pendengar.

## **Heatmap**

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data=df[['popularity', 'duration_ms']].corr(),
            annot=True,
            cmap='RdPu_r',
            fmt='.2f')
plt.title('Korelasi Antar Variabel Numerik Lagu')
plt.show()
```

Heatmap menunjukkan bahwa korelasi antara popularitas dan durasi lagu sangat lemah (sekitar 0.02). Hal ini menandakan bahwa panjang lagu hampir tidak memiliki pengaruh terhadap tingkat popularitasnya.

Melihat korelasi yang sangat rendah tersebut, analisis selanjutnya dapat memasukkan variabel lain yang lebih berpotensi mempengaruhi popularitas lagu, seperti tempo, energy, danceability, atau strategi promosi. Dengan mempertimbangkan lebih banyak faktor, model analisis dapat memberikan pemahaman yang lebih akurat mengenai apa yang membuat sebuah lagu menjadi populer.
